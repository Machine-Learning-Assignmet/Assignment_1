```python
# this is to demonstrate a simple import statement

import pandas as pd
```

```python
df = pd.read_csv("loan.csv")
df.head()
```

```python
selected_columns = [
    "loan_amnt", "member_id", "funded_amnt", "funded_amnt_inv", "term",
    "int_rate", "installment", "emp_length", "home_ownership", "verification_status",
    "pymnt_plan", "purpose", "addr_state", "dti", "delinq_2yrs",
    "earliest_cr_line", "open_acc", "pub_rec", "total_acc",
    "out_prncp", "total_rec_int", "total_rec_late_fee", "last_pymnt_d",
    "application_type", "annual_inc_joint", "acc_now_delinq", "tot_coll_amt",
    "tot_cur_bal", "open_acc_6m", "total_bal_il", "loan_status"
]

df = df[selected_columns]

df.head()
```

```python
# Fill missing 'annual_inc_joint' with 0 for individual applicants
df.loc[df['application_type'] == 'INDIVIDUAL', 'annual_inc_joint'] = df.loc[
    df['application_type'] == 'INDIVIDUAL', 'annual_inc_joint'
].fillna(0)

df['annual_inc_joint'].unique()
```

```python
# Convert 'term' from string to integer
df['term'] = df['term'].str.replace(' months', '').astype(int)
df['term'].unique()
```

```python
# Step 1: Replace special cases, leave NaN untouched
df['emp_length'] = df['emp_length'].replace({
    '< 1 year': 0,
    '10+ years': 10
})

# Step 2: Remove ' years' or ' year' from remaining strings
# Only affects strings, numeric values are untouched
df['emp_length'] = df['emp_length'].astype(str).str.replace(' years', '').str.replace(' year', '')

# Step 3: Convert everything to numeric, coercing non-convertible to NaN
df['emp_length'] = pd.to_numeric(df['emp_length'], errors='coerce')

df['emp_length'].unique()
```

```python
# Convert columns to numeric automatically
# For consistency, use lower cases and convert any to other

# Convert all values to lowercase
df['home_ownership'] = df['home_ownership'].str.lower()

# Replace 'any' and 'none' with 'other'
df['home_ownership'] = df['home_ownership'].replace({
    'any': 'other'
})

df['home_ownership'], _ = pd.factorize(df['home_ownership'])

# For payment plan
df['pymnt_plan'] = df['pymnt_plan'].replace({
    'n' : 0,
    'y' : 1
})

# For purpose
df['purpose'], _ = pd.factorize(df['purpose'])


# For address state
df['addr_state'], _ = pd.factorize(df['addr_state'])

df[['home_ownership','pymnt_plan','purpose','addr_state']].head()

```

```python
# Binary Conversions
df['verification_status'] = df['verification_status'].replace({
    'Not Verified': 0,
    'Verified': 1,
    'Source Verified': 1   #Who verified does not really matter per online research
})



# For application type
df['application_type'] = df['application_type'].str.lower()

df['application_type'] = df['application_type'].replace({
    'individual' : 0,
    'joint' : 1
})
df[['verification_status','application_type']].head()
```

```python
# Convert 'earliest_cr_line' to datetime
df['earliest_cr_line'] = pd.to_datetime(df['earliest_cr_line'], format='%b-%y', errors='coerce')
df['earliest_cr_line'] = df['earliest_cr_line'].dt.strftime('%m-%Y')



# Convert 'last_pymnt_d' to datetime
df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], format='%b-%y', errors='coerce')
df['last_pymnt_d'] = df['last_pymnt_d'].dt.strftime('%m-%Y')


df[['earliest_cr_line','last_pymnt_d']].head()
```

```python
# Count missing values per column
missing_counts = df.isna().sum()

# Show only columns that have at least one missing value
missing_columns = missing_counts[missing_counts > 0]
print(missing_columns)
```

```python
cols_missing_29 = ['delinq_2yrs', 'earliest_cr_line', 'open_acc', 
                   'pub_rec', 'total_acc', 'acc_now_delinq']

df[df[cols_missing_29].isna().all(axis=1)].shape

df = df.dropna(subset=cols_missing_29, how='all').reset_index(drop=True)

# Confirm they’re gone
print(df[cols_missing_29].isna().sum())
```
